```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset("tips")
df
```

### Seaborn is a Python data visualization library based on matplotlib.It provides a high_level interface for drawing attractive and informative statistical graphics.

### Numerical Plots

### 1.Distribution plot

```python
sns.distplot(x=df["total_bill"],color="darkcyan",bins=10)
```

### 2.Joint_plot

```python
sns.jointplot(y="tip",x="total_bill",data=df,kind="reg")
```

```python
sns.jointplot(y="tip",x="total_bill",data=df,kind="hex",color="darkcyan")
```

```python
sns.jointplot(y="tip",x="total_bill",data=df,kind="kde")
```

```python
sns.kdeplot(y="tip",x="total_bill",data=df)
```

### 4.Pair_plot

```python
sns.pairplot(df,hue="sex",palette="Blues")
```

```python
sns.rugplot(df["total_bill"])
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))
sns.jointplot(y="tip",x="total_bill",data=df,kind="reg")
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8,6))
sns.jointplot(y="tip",x="total_bill",data=df,kind="reg")
```

```python
sns.set_style("ticks")
plt.figure(figsize=(8,6))
sns.set_context("paper",font_scale=1.5)
sns.jointplot(y="tip",x="total_bill",data=df,kind="reg")
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))
sns.jointplot(y="tip",x="total_bill",data=df,kind="reg")
sns.despine(left=False)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,6))
sns.jointplot(y="tip",x="total_bill",data=df,kind="reg")
sns.despine(left=True)
```

```python
sns.relplot("total_bill","tip",data=df,style="smoker",hue="sex",)
```

### Categorical Plots

### Bar plot


```python
sns.barplot(x="smoker",y="total_bill",data=df,estimator=np.mean)
```

```python
sns.barplot(x="sex",y="total_bill",data=df,estimator=np.std)
```

```python
sns.countplot(x="smoker",data=df)
```

### Boxplot

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=df)
plt.legend(loc=0)
```

### Violin_plt

### It is the combination of Box_plot and kde_plot

```python
sns.violinplot(x="day",y="total_bill",hue="sex",data=df,split=True)
```

```python
sns.violinplot(x="day",y="total_bill",hue="sex",data=df)
```

### Strip_Plot

    It is used to show the correlation for one categorical and numerical variable

```python
sns.stripplot(x="day",y="total_bill",data=df)
```

```python
sns.stripplot(x="day",y="total_bill",data=df,jitter=True)
```

```python
sns.stripplot(x="day",y="total_bill",data=df,jitter=True,hue="sex")
```

```python
sns.stripplot(x="day",y="total_bill",data=df,jitter=True,hue="sex",dodge=True)
```

### Swarm_Plot

```python
sns.swarmplot(x="day",y="total_bill",data=df,hue="sex")
```

```python
sns.swarmplot(x="day",y="total_bill",data=df,hue="sex",dodge=True)
```

```python
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",dodge=True,palette="magma")
sns.set_style('whitegrid')
sns.set_context('paper')
```

```python
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",dodge=True,palette="viridis")
sns.set_style('whitegrid')
sns.set_context('paper')
```

```python
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",dodge=True,palette="afmhot")
sns.set_style('darkgrid')
sns.set_context('poster')
plt.legend(loc=0)
```
